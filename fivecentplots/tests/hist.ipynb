# <b>hist</b>

The `hist` function allows histogram visualizing DataFrame data.  At a minimum, the `hist` function requires the following keywords:
<ul>
<li> `df`: a pandas DataFrame</li>
<li> `y`: the name of the DataFrame column containing the y-axis data </li>
</ul>

Other optional keywords are described in <a href="https://endangeredoxen.github.io/fivecentplots/keyword.html">Keyword Arguments</a>.

### Other

```python
SHOW = False
```

## Simple histogram

### Vertical bars

A simple histogram with default bin size of 20:

```python
fcp.hist(df=df, x='Value', show=SHOW)
```

### Horizontal bars

Same data as above but with histogram bars oriented horizontally:

```python
fcp.hist(df=df, x='Value', show=SHOW, horizontal=True)
```

## Legend

Add a legend:

```python
fcp.hist(df=df, x='Value', show=SHOW, legend='Region')
# ymax limit is wrong
```

## Kernal density estimator

```python
fcp.hist(df=df, x='Value', show=SHOW, legend='Region', kde=True, kde_width=2)
```

## Row/column plot

Make multiple subplots with different row/column values:

```python
fcp.hist(df=df, x='Value', show=SHOW, legend='Region', col='Batch', row='Sample', kde=True, ymax=10)
# yaxis limit and values make no sense with or without kde
# labels not swapping when horizontal
# kde is giving weird normalized counts
```

## Wrap plot

```python
fcp.hist(df=df, x='Value', show=SHOW, legend='Region', wrap='Batch', ax_size=[250, 250])  # , kde=True, kde_width=2,
# same comment as above
# ytick label spacing is wrong with kde but looks ok without
# why do we have non-integer ticks at all?
```
